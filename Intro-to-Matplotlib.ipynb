### Two options for subplots

```python
#Option 1
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,
                                       ncols=2,
                                       figsize=(10,5))
#Plot for each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot Option 2
fig,ax=plt.subplots(nrows=2, ncols=2,figsize=(10,5))
#Plot for each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

### Which one should we use?(Pyplot vs Matplotlib OO)
* When plotting something quickly, Okay to use pyplot method
* When something more advanced, use the OO method

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
len(over_50)
```

```python
#Pyplot method
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target");
```

```python
#OO method mixed with pyplot method
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax);
ax.set_xlim([45,100]);
```

```python
# OO METHOD(ASLI WALA)
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"]);
#Customise the Plot
ax.set(title="Heart Disease and Cholesterol levels",
       xlabel="AGE",
       ylabel="CHOLESTEROL");
#ADD A LEGEND
ax.legend(*scatter.legend_elements(),title="Target");
#Add a Horizontal Line
ax.axhline(over_50["chol"].mean(),linestyle='--');
```

```python
heart_disease.head()
```

```python
fig,(ax1,ax2)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)
scatter=ax1.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);
scatter2=ax2.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["slope"]);
ax1.legend(*scatter.legend_elements(),title="Target");
ax2.legend(*scatter2.legend_elements(),title="Slope");
ax1.set(title="AGE AND CHOLESTEROL",
        ylabel="CHOLESTEROL");
ax2.set(title="AGE AND MAX HEART RATE",
        xlabel="AGE",
        ylabel="THALACH");
ax1.axhline(over_50["chol"].mean(),linestyle='-.');
ax2.axhline(over_50["thalach"].mean(),linestyle=':');
fig.suptitle("Heart Disease Analysis",fontsize=20,fontweight="bold");
```

```python
#set the style
plt.style.use('seaborn-v0_8')
# OO METHOD(ASLI WALA)
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"],
                  cmap="winter");
#Customise the Plot
ax.set(title="Heart Disease and Cholesterol levels",
       xlabel="AGE",
       ylabel="CHOLESTEROL");
#ADD A LEGEND
ax.legend(*scatter.legend_elements(),title="Target");
#Add a Horizontal Line
ax.axhline(over_50["chol"].mean(),linestyle='--');
```

```python
plt.style.use('seaborn-v0_8-darkgrid')
fig,(ax1,ax2)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)
scatter=ax1.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                   cmap="plasma");
scatter2=ax2.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["slope"],
                    cmap="winter");
ax1.legend(*scatter.legend_elements(),title="Target");
ax2.legend(*scatter2.legend_elements(),title="Slope");
ax1.set(title="AGE AND CHOLESTEROL",
        ylabel="CHOLESTEROL");
ax1.set_xlim([50,80]);
ax2.set(title="AGE AND MAX HEART RATE",
        xlabel="AGE",
        ylabel="THALACH");
ax2.set_xlim([50,80]);
ax2.set_ylim([100,200]);
ax1.axhline(over_50["chol"].mean(),linestyle='-');
ax2.axhline(over_50["thalach"].mean(),linestyle='--');
fig.suptitle("Heart Disease Analysis",fontsize=20,fontweight="bold");
```
